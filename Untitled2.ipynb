```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Cargar el conjunto de datos
data = pd.read_csv('agrop.csv')

# Función para convertir cadenas formateadas a flotantes
def convert_to_float(x):
    if isinstance(x, str):
        # Eliminar espacios y comas
        return float(x.replace(' ', '').replace(',', ''))
    return x

# Definir las características numéricas y categóricas
numeric_features = ['Área sembrada', 'Área cosechada', 'Producción', 'Rendimiento']
categorical_features = ['Grupo cultivo', 'Cultivo', 'Estado físico del cultivo', 'Nombre científico del cultivo']

# Aplicar la función a las columnas numéricas
for feature in numeric_features:
    data[feature] = data[feature].apply(convert_to_float)

# Crear transformadores para las características numéricas y categóricas
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Crear el preprocesador con los transformadores
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Aplicar las transformaciones definidas al conjunto de datos
transformed_data = preprocessor.fit_transform(data)

# Convertir los datos transformados a un DataFrame de pandas
transformed_df = pd.DataFrame(transformed_data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Cargar el conjunto de datos
data = pd.read_csv('agrop.csv')

# Eliminar columnas con todos los valores faltantes
data = data.dropna(axis=1, how='all')

# Imputar valores NaN en 'Rendimiento' con la mediana
data['Rendimiento'] = data['Rendimiento'].fillna(data['Rendimiento'].median())

# Identificar características categóricas y numéricas
categorical_features = ['Grupo cultivo', 'Cultivo', 'Estado físico del cultivo', 'Nombre científico del cultivo']
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_features.remove('Rendimiento')  # Eliminar la variable objetivo de las características numéricas

# Crear transformadores para las características numéricas y categóricas
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())  # Escalado de características
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # Codificación one-hot
])

# Crear el preprocesador con los transformadores
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Preprocesar las características
X = data.drop('Rendimiento', axis=1)  # Características
y = data['Rendimiento']  # Variable objetivo
X_preprocessed = preprocessor.fit_transform(X)

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.19, random_state=36)

# Inicializar y entrenar el modelo de regresión lineal
regression_model = Ridge(alpha=8.0)
regression_model.fit(X_train, y_train)

# Evaluar el rendimiento del modelo de regresión lineal
linear_reg_score = regression_model.score(X_test, y_test)
linear_reg_predictions = regression_model.predict(X_test)
linear_reg_mse = mean_squared_error(y_test, linear_reg_predictions)
linear_reg_r2 = r2_score(y_test, linear_reg_predictions)

# Imprimir los resultados del modelo de regresión lineal
print("Resultados de la regresión lineal:")
print("Score:", linear_reg_score)
print("Mean Squared Error:", linear_reg_mse)
print("R2 Score:", linear_reg_r2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Cargar el conjunto de datos
data = pd.read_csv('agrop.csv')

# Imputar valores NaN en 'Rendimiento' con la mediana
data['Rendimiento'] = data['Rendimiento'].fillna(data['Rendimiento'].median())

# Identificar características categóricas y numéricas
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_features.remove('Rendimiento')  # Eliminar la variable objetivo de las características numéricas

# Crear el preprocesador con transformadores para características numéricas y categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),  # Imputación de valores faltantes
            ('scaler', StandardScaler())  # Escalado de características
        ]), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Preprocesar las características
X = data.drop('Rendimiento', axis=1)  # Características
y = data['Rendimiento']  # Variable objetivo
X_preprocessed = preprocessor.fit_transform(X)

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.19, random_state=34)

# Inicializar el modelo de regresión polinomial
poly_reg = PolynomialFeatures(degree=3)

# Transformar las características a características polinomiales
X_poly_train = poly_reg.fit_transform(X_train)
X_poly_test = poly_reg.transform(X_test)

# Entrenar el modelo de regresión polinomial
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_poly_train, y_train)

# Evaluar el rendimiento del modelo de regresión polinomial
poly_reg_score = poly_reg_model.score(X_poly_test, y_test)
poly_reg_predictions = poly_reg_model.predict(X_poly_test)
poly_reg_mse = mean_squared_error(y_test, poly_reg_predictions)
poly_reg_r2 = r2_score(y_test, poly_reg_predictions)

# Imprimir los resultados del modelo de regresión polinomial
print("Resultados del modelo de regresión polinomial:")
print("Score:", poly_reg_score)
print("Mean Squared Error:", poly_reg_mse)
print("R2 Score:", poly_reg_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
import pandas as pd

# Cargar el conjunto de datos
data = pd.read_csv('agrop.csv')

# Imputar valores NaN en 'Rendimiento' con la mediana
data['Rendimiento'] = data['Rendimiento'].fillna(data['Rendimiento'].median())

# Identificar características categóricas y numéricas
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_features.remove('Rendimiento')  # Eliminar la variable objetivo de las características numéricas

# Crear el preprocesador con transformadores para características numéricas y categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='median'), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Preprocesar las características
X = data.drop('Rendimiento', axis=1)  # Características
y = data['Rendimiento']  # Variable objetivo
X_preprocessed = preprocessor.fit_transform(X)

# Asegurarse de que no hay valores NaN en y antes de dividir los datos
y = y.fillna(y.median())

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=36)

# Inicializar el modelo de Random Forest
random_forest_reg = RandomForestRegressor(n_estimators=50, random_state=36)

# Entrenar el modelo de Random Forest
random_forest_reg.fit(X_train, y_train)

# Evaluar el rendimiento del modelo de Random Forest
random_forest_reg_score = random_forest_reg.score(X_test, y_test)

# Realizar validación cruzada para evaluar la robustez del modelo
random_forest_reg_cv_scores = cross_val_score(random_forest_reg, X_preprocessed, y, cv=5)

# Realizar predicciones
random_forest_reg_predictions = random_forest_reg.predict(X_test)

# Calcular métricas de desempeño adicionales (por ejemplo, error cuadrático medio y coeficiente de determinación)
random_forest_reg_mse = mean_squared_error(y_test, random_forest_reg_predictions)
random_forest_reg_r2 = r2_score(y_test, random_forest_reg_predictions)

# Imprimir los resultados del modelo de Random Forest
print("Resultados de Random Forest:")
print("Score:", random_forest_reg_score)
print("Cross-Validation Scores:", random_forest_reg_cv_scores)
print("Mean Squared Error:", random_forest_reg_mse)
print("R2 Score:", random_forest_reg_r2)
```
